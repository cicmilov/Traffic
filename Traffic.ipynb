## In next cell we will group data by "Event Parameter". Then we will look for any irregular order of events, and print them.
This is done by checking in every group every code 1 and its next 5 rows in table if they are in regular order.
Also for every other code number we check few previuos and few next if they are in regular order

```python
group = df.groupby("Event Parameter")
yellow_time_dict = {}
red_time_dict = {}
for ep, tab in group:    
    i = 0
    yellow_time = []
    red_time = []
    while i < len(tab)-6:
        ec = tab.iloc[i]["Event Code"]
        if ec == 1:
            if (tab.iloc[i:i+6]["Event Code"] == [1, 7, 8, 9, 10, 11]).sum() < 6:
                print(tab.iloc[i:i+6])
                print("*******************************")
                print(tab.index[i])
                i += 5
            else:
                yellow_time.append(tab.iloc[i+3]["Event Time"] - tab.iloc[i+2]["Event Time"])
                red_time.append(tab.iloc[i+5]["Event Time"] - tab.iloc[i+4]["Event Time"])
        else:
            if (tab.iloc[i-ec+6:i-ec+12]["Event Code"] == [1, 7, 8, 9, 10, 11]).sum() < 6:
                print(tab.iloc[i-ec+6:i-ec+12])
        i += 1
    yellow_time_dict[ep] = np.median(np.array(yellow_time))
    red_time_dict[ep] = np.median(np.array(red_time))
    print(ep, "median yellow time is ", np.median(np.array(yellow_time)))
    print(ep, "median red time is ", np.median(np.array(red_time)))
        
```

We can see in result of cell above that errors are badly orederd phases of traffic light, so we will fix that in our main dataframe.
Each error has 4 timestamps, which represent all 6 events, just in wrong order. So we will rewrite data in original table in right way.

```python
def correct_data(df, tab, i, ep):
    times = sorted(list(set(tab.iloc[i:i+6]["Event Time"])))
    df.loc[tab.index[i]] = [times[0], 1, "Phase Begin Green", ep] 
    df.loc[tab.index[i+1]] = [times[1], 7, "Phase Green Termination", ep] 
    df.loc[tab.index[i+2]] = [times[1], 8, "Phase Begin Yellow Clearance", ep] 
    df.loc[tab.index[i+3]] = [times[2], 9, "Phase End Yellow Clearance", ep] 
    df.loc[tab.index[i+4]] = [times[2], 10, "Phase Begin Red Clearance", ep] 
    df.loc[tab.index[i+5]] = [times[3], 11, "Phase End Red Clearance", ep] 
```

```python
for ep, tab in group:    
    i = 0
    while i < len(tab)-6:
        ec = tab.iloc[i]["Event Code"]
        # we only do this for event code 1, because all the errors are started with this value
        if ec == 1:
            if (tab.iloc[i:i+6]["Event Code"] == [1, 7, 8, 9, 10, 11]).sum() < 6:
                correct_data(df, tab, i, ep)
        i += 1
```

### Next we will look for same values of time stamps in 'Phase Green Termination' and 'Phase Begin Yellow Clearance' and 'Phase End Yellow Clearance' and 'Phase Begin Red Clearance'

```python
group = df.groupby("Event Parameter")
for ep, tab in group:    
    i = 0
    while i < len(tab)-6:
        ec = tab.iloc[i]["Event Code"]
        if ec == 1:
            if tab.iloc[i+1]["Event Time"] != tab.iloc[i+1]["Event Time"]:
                print(tab.iloc[i:i+6])
                print("'Phase Green Termination' and 'Phase Begin Yellow Clearance'")
            if tab.iloc[i+3]["Event Time"] != tab.iloc[i+4]["Event Time"]:
                print(tab.iloc[i:i+6])
                print("'Phase End Yellow Clearance' and 'Phase Begin Red Clearance'")
        i += 1
```

### Last thing we will check time passed between 'Phase End Red Clearance' and 'Phase Begin Green'

```python
time_pessed = []
for i in range(len(df)):
    if df.iloc[i]["Event Code"] == 10:
        j = i
        while df.iloc[j]["Event Code"] != 11 or df.iloc[j]["Event Parameter"] != df.iloc[j]["Event Parameter"]:
            if df.iloc[j]["Event Code"] == 1:
                time_green_start = df.iloc[j]["Event Time"]
            j += 1
        time_red_end = df.iloc[j]["Event Time"]
        time_pessed.append(time_red_end - time_green_start)
        
time_p = np.array(time_pessed)
print(time_p.mean())
print(time_p.max())
print(time_p.min())
```
